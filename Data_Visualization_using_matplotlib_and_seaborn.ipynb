```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars_data=pd.read_csv('Toyota.csv')
```

```python
#to remove missing values from dataframe
cars_data.dropna(axis=0, inplace=True) #if axis=0 then we are considering rows and if axis=1 then columns
```

```python
#scatter plot is use to represent correlation between two variables
plt.scatter(cars_data['Age'], cars_data['Price'], c='red')
plt.title('Scatter plot of Price vs Age of cars')
plt.xlabel('Age (months)')
plt.ylabel('Price (Euros)')
plt.show()
```

```python
#Histogram is use to represent the frequency distribution of numerical variables
plt.hist(cars_data['KM'],
        color= 'green',
        edgecolor= 'white',
        bins=5)
plt.title('Histogram of Kilometer')
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
```

```python
#Bar plot
#to represent the frequency distribution of categorical variables
#A bar diagram makes it easy to compare sets of data between different groups
counts=cars_data['FuelType'].value_counts()
fuelType=cars_data['FuelType'].unique()
index=np.arange(len(fuelType))

plt.bar(index, counts, color=['r','b','g'])
plt.title('Bar Plot of Fuel Types')
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.xticks(index,fuelType,rotation=0)
plt.show()
```

# Seaborn library for visualization

```python
import seaborn as sns
```

```python
sns.set(style='darkgrid')
sns.regplot(x=cars_data['Age'], y=cars_data['Price'])
```

```python
sns.set(style='darkgrid')
sns.regplot(x=cars_data['Age'], y=cars_data['Price'], fit_reg=False,marker='*')
```

```python
sns.set(style='darkgrid')
sns.lmplot(x='Age', y='Price', data=cars_data, fit_reg=False, hue='FuelType', legend=True, palette='Set1')
#sns.title('Scatter Plot of Price vs Age byFuelType')
```

```python
#Histogram gives kernel density estimate
sns.distplot(cars_data['Age'])
```

```python
#Histogram without kernel density estimate
sns.distplot(cars_data['Age'], kde=False)
```

```python
#In above histogram many bins are there, so we are reducing bins
sns.distplot(cars_data['Age'],kde=False, bins=5)
```

```python
#Bar plot is use to show frequency distribution of any categorical variable
sns.countplot(x='FuelType', data=cars_data)
```

```python
#Grouped bar plot of FuelType and Automatic
sns.countplot(x='FuelType', data=cars_data, hue='Automatic')
```

# Box and Whiskers plot - numerical data
## Box and whiskers plot of Price to visually interpret the five-number summary: min,max and 3 quantile

```python
sns.boxplot(y=cars_data['Price'])
```

```python
#Box and whiskers plot for numerical vs categorical variable
sns.boxplot(x=cars_data['FuelType'], y=cars_data['Price'])
```

```python
# Grouped box and whiskers plot of Price vs FuelType and Automatic
sns.boxplot(x='FuelType', y='Price', hue='Automatic', data=cars_data)
```

```python
f, (ax_box, ax_hist)=plt.subplots(2, gridspec_kw={"height_ratios": (0.15,0.85)})
sns.boxplot(cars_data['Price'], ax=ax_box)
sns.distplot(cars_data['Price'], ax=ax_hist, kde=False)
```
